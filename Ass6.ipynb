```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('forestfires.csv')
df
```

# 1) Histplot

```python
sns.histplot(df, x='wind', hue='X')
plt.title("Histplot")
```

# 2) Box plot

```python
sns.boxenplot(df, x="temp")
```

# 3) Single line graph

```python
sns.lineplot(df, x="X", y="Y")
```

# 4) Multi line graph 

```python
sns.lineplot(df, x="X", y="Y", hue="day")
```
